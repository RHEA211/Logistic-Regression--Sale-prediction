```python
#importing libraries
import pandas as pd
import numpy as np
```

```python
#loading the dataset using its path
dataset= pd.read_excel(r'C:\Users\KIIT\Desktop\Sale.xlsx')
dataset
```

```python
#getting an idea of the data
print(dataset.shape)
print(dataset.head(10))
```

```python
#segragating the dataset into X and Y
X=dataset.iloc[:,:-1].values
X
```

```python
#segragating the dataset into X and Y
Y=dataset.iloc[:,-1].values
Y
```

```python
#Splitting the dataset inte train and test
#Here 25% of the dataset is being used for test, 75% for train

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.25, random_state = 0)
```

```python
#feature scaling done so that the model is not biased towards larger values

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train) 
X_test = sc.transform(X_test) 
```

```python
#Training the model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state = 0)
model.fit(X_train, y_train)
```

```python
#Final prediction
age = int(input("Enter Age of Customer: "))
sal = int(input("Enter Salary Of customer: "))
newCust = [[age,sal]]
result = model.predict(sc.transform(newCust))
print(result)
if result == 1:
  print("Customer will Buy")
else:
  print("Customer won't Buy")
```

```python
#Prediction for the test data(actual v/s predicted)
y_pred = model.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
#Using Confusion Matrix and finding out accurcy of the model
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix: ")
print(cm)

print("Accuracy of your Model is: {0}%".format(accuracy_score(y_test, y_pred)*100))
```
